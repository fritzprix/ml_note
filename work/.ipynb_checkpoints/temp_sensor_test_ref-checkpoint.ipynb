     * Gradient
     * Amount of drifted TMU & Temperature releation in running state transition
```
```
* Significant outcome
  * It seems there is H/W characteristic variation which causes significant difference in TMU vs. Temperature relation.
    * Potential causes for the difference might be...
      * Thermal Coupling between Temperature sensor and x220 SoC due to...
        * PCB (Thermal conductivity variation)
        * PBA assembly process variation (amount of solder paste, reflow condition)
      * Cooling efficiency of heatsink due to..
        * Thermal coupling characteristic variation in assembly process
  * Drift in TMU vs. Temperature is observed from both test targets 
     * It's clear transition in running state of SoC causes the drift in TMU vs. Temperature. more specifically, it can be..
       * Voltage drop (DC Component) caused by parasitic resistance between supply and load (x220)
     * **Drift might be compensated by simple linear regression model trained with additional data which represents load condition**
  * Variance of TMU for given Temperature is getting larger when temperature rises. 
     * Voltage noise level (AC Component)

## Load adaptive TMU compensation model
* Challenge from TMU vs Temperature drift from running test
  * TMU vs. Temperature has at least 2 intersection points (1 : 2) in simple TMU vs Temperature relation depending on running state of AI core
  * Through many test cases observed so far, it seems to be clear that load condition (how much power consumed by AI core) is related to TMU vs Temperature relation.
  * loewr supply voltage, lower TMU value 
  * **key assumption : supply voltage variation is dominant factor for the drift**

* Approach
  * Simple linear regression model trained with only TMU data as input showed limitation on prediction performance due to TMU drift
  * develop new linear regression model takes not only TMU but also supply voltage condition (including current) as input feature.
  * training data (TMU / Temperature / supply voltage / current) is collected from H2 board in its idle state with varying supply voltage from 1.00V ~ 1.04V 
  * validation data 
    * #1 1.05V idle test data (from the board from which the training data is collected)
    * #2 1.00V idle test data (from the other board)
    * #3 1.00V run test data
    
* Result 
  * Model shows fairly decent performance for #1 case (RMSE Score 1.83)
  * Bad performance for #2 & #3 case (RMSE Score 31.1 & RMSE Score 172.05)

````python
v_meas_data_h2_b1_0 = pd.read_csv('/home/data/power_measure_0.csv')
v_meas_data_h2_b1_1 = pd.read_csv('/home/data/power_measure_1.csv')

v_meas_data_h2_b1_0["c_deg"] = v_meas_data_h2_b1_0["tsens0"] / 2
v_meas_data_h2_b1_1["c_deg"] = v_meas_data_h2_b1_1["tsens0"] / 2

v_meas_data_h2_b1_0["tmu pred"] = temp_vs_tmu_reg.predict(np.c_[v_meas_data_h2_b1_0["c_deg"]])
v_meas_data_h2_b1_1["tmu pred"] = temp_vs_tmu_reg.predict(np.c_[v_meas_data_h2_b1_1["c_deg"]])

v_meas_data_h2_b1_0["tmu_diff_from_vdrop"] = v_meas_data_h2_b1_0["tmu pred"] - v_meas_data_h2_b1_0["tmu.avg"]
v_meas_data_h2_b1_1["tmu_diff_from_vdrop"] = v_meas_data_h2_b1_1["tmu pred"] - v_meas_data_h2_b1_1["tmu.avg"]

v_meas_data_h2_b1_0["vdrop_pred"] = v_meas_data_h2_b1_0["tmu_diff_from_vdrop"] / v_vs_tmu_reg.coef_ * 1000
v_meas_data_h2_b1_1["vdrop_pred"] = v_meas_data_h2_b1_1["tmu_diff_from_vdrop"] / v_vs_tmu_reg.coef_ * 1000

v_meas_data_h2_b1_0["esr"] = v_meas_data_h2_b1_0["vdrop_pred"] / v_meas_data_h2_b1_0["core.i"]
v_meas_data_h2_b1_1["esr"] = v_meas_data_h2_b1_1["vdrop_pred"] / v_meas_data_h2_b1_1["core.i"]

v_meas_data_h2_b1_0["v_avg"] = wnd_avg(v_meas_data_h2_b1_0["core.v"], 50)
v_meas_data_h2_b1_1["v_avg"] = wnd_avg(v_meas_data_h2_b1_1["core.v"], 50)


a = 0.5

fig, ax = plt.subplots(6,1,figsize=(15,25))
ax[0].scatter(x=v_meas_data_h2_b1_0["elapsed sec."],y=v_meas_data_h2_b1_0["tmu pred"], label="TMU Predicted from Temp sensor #0", alpha=a)
ax[0].scatter(x=v_meas_data_h2_b1_0["elapsed sec."],y=v_meas_data_h2_b1_0["tmu.avg"], label="TMU Measured #0", alpha=a)
ax[0].scatter(x=v_meas_data_h2_b1_1["elapsed sec."],y=v_meas_data_h2_b1_1["tmu pred"], label="TMU Predicted from Temp sensor #1", alpha=a)
ax[0].scatter(x=v_meas_data_h2_b1_1["elapsed sec."],y=v_meas_data_h2_b1_1["tmu.avg"], label="TMU Measured #1", alpha=a)
ax[0].set_title("TMU measured vs predicted")
ax[0].legend()
ax[1].scatter(x=v_meas_data_h2_b1_0["elapsed sec."], y=v_meas_data_h2_b1_0["vdrop_pred"], label="Voltage drop[mV] #0")
ax[1].scatter(x=v_meas_data_h2_b1_1["elapsed sec."], y=v_meas_data_h2_b1_1["vdrop_pred"], label="Voltage drop[mV] #1")
ax[1].set_title("Voltage drop (predicted based on TMU change) [mV]")
ax[1].legend()
ax[2].scatter(x=v_meas_data_h2_b1_0["elapsed sec."], y=v_meas_data_h2_b1_0["core.i"] / 1000, label="core.i #0")
ax[2].scatter(x=v_meas_data_h2_b1_1["elapsed sec."], y=v_meas_data_h2_b1_1["core.i"] / 1000, label="core.i #1")
ax[2].set_title("Output current from VRM(Flex BMR469) [A]")
ax[2].set_ylim(0, 20)
ax[2].legend()
ax[3].scatter(x=v_meas_data_h2_b1_0["elapsed sec."], y=v_meas_data_h2_b1_0["esr"], label="ESR #0")
ax[3].scatter(x=v_meas_data_h2_b1_1["elapsed sec."], y=v_meas_data_h2_b1_1["esr"], label="ESR #1")
ax[3].set_title("Calculated ESR (Equivalent Serial Resistance) [Ohm] from predicted voltage drop & measured output current")
ax[3].legend()
ax[4].plot(v_meas_data_h2_b1_0["elapsed sec."], v_meas_data_h2_b1_0["core.v"] / 1000, label="core.v #0")
ax[4].plot(v_meas_data_h2_b1_1["elapsed sec."], v_meas_data_h2_b1_1["core.v"] / 1000, label="core.v #1")
ax[4].set_title("Output voltage from VRM(Flex BMR469) [V]")
ax[4].legend()
ax[5].scatter(x=v_meas_data_h2_b1_0["elapsed sec."], y=v_meas_data_h2_b1_0["core.duty"] / 1000, label="core.duty #0")
ax[5].scatter(x=v_meas_data_h2_b1_1["elapsed sec."], y=v_meas_data_h2_b1_1["core.duty"] / 1000, label="core.duty #1")
ax[5].set_title("Switching Duty @ VRM [%]")
ax[5].legend()
plt.show()








````
