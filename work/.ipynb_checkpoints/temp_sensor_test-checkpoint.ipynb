## Forced H/W failure test 
* By running bert.batch7 test on the H2 board which heatsink is removed from, H/W failure can be reproduced successfully through all 6 trials. test logs were broken for 4 out of 6 trial, so valid data collected from just 2 out of 6. trace player was terminated with printing error message below for both cases.

```C
trace_player_opt.cc:609] Check failed: dcs_status.checksum_error == 0 (1 vs. 0) Checksum Error!
```
* Test is composed of two consecutive trace player executions. the first run BERT (batch 7) to stress H/W and the other keeps it in idle to give H/W cool down cycle
* Discontinuity on the both cases is due to execution delay between consecutive executions of trace player (load command and data into board memeory, etc.) 
* Significant outcome of test
  * System reboot (maybe due to PCI drop) happens during cool down phase of test 
     * Temperature sensor and TMU value indicates fairly safe temperature range though..
     * Does that imply there is any unknown source of cuase for the failure?
  * Temperature (or TMU) values related to trace player abnormal termination are different for each case
     * H/W characteristic varies widely, it seems that every H/W has different failure condition

## Load adaptive TMU compensation model
* Challenge from TMU vs Temperature drift from running test
  * TMU vs. Temperature has at least 2 intersection points (1 : 2) in simple TMU vs Temperature relation depending on running state of AI core
  * Through many test cases observed so far, it seems to be clear that load condition (how much power consumed by AI core) is related to TMU vs Temperature relation.
  * loewr supply voltage, lower TMU value 
  * **key assumption : supply voltage variation is dominant factor for the drift**

* Approach
  * Simple linear regression model trained with only TMU data as input showed limitation on prediction performance due to TMU drift
  * develop new linear regression model takes not only TMU but also supply voltage condition (including current) as input feature.
  * training data (TMU / Temperature / supply voltage / current) is collected from H2 board in its idle state with varying supply voltage from 1.00V ~ 1.04V 
  * validation data 
    * #1 1.05V idle test data (from the board from which the training data is collected)
    * #2 1.00V idle test data (from the other board)
    * #3 1.00V run test data
    
* Result 
  * Model shows fairly decent performance for #1 case (RMSE Score 1.83)
  * Bad performance for #2 & #3 case (RMSE Score 31.1 & RMSE Score 172.05)
